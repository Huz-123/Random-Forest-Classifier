```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_digits
```

```python
dig= load_digits()
```

```python
dir(dig)
```

```python
plt.gray()
for i in range(10):
  plt.matshow(dig.images[i])
```

```python
dig.data[:5]
```

```python
df = pd.DataFrame(dig.data)
```

```python
df.head()
```

```python
dig.target[:20]
```

```python
df['target'] = dig.target
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.drop(['target'], axis='columns'), dig.target, test_size=0.2)
```

```python
len(x_train), len(x_test), len(y_train), len(y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier  # ensemble is tearm that used when multiple model used to predict the outcome
model= RandomForestClassifier(n_estimators = 10000)  # best use for 100 to 200
model.fit(x_train, y_train)                          # when i use 200 score(0.977777777 )
```

```python
model.score(x_test, y_test)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
sns.heatmap(cm, annot= True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
